```python
df = sns.load_dataset("titanic")
df

# toplulaştırma & gruplama (aggregation & grouping)
```

```python
df["age"].mean()

# aggregation örneği
```

```python
df.groupby("sex")["age"].mean()

# groupby ile grouping yapıp aggregation yaptık
```

```python
df.groupby("sex").agg({"age":"mean"})

# groupby ile grupladık. .agg ile içine dict atıp hangi değişkene ne yapacağını belirttik
```

```python
df.groupby("sex").agg({"age" : ["mean","sum"]})

# yaptığımız işlemi çoğaltabiliriz
```

```python
df.groupby("sex").agg({"age" : ["mean","sum"], "embark_town" : "count"})
```
