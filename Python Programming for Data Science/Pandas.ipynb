```python
df = sns.load_dataset("titanic")
# apply & lambda 
```

```python
df["age2"] = df["age"]*2
df["age3"] = df["age"]*5
df.head()
```

```python
for col in df.columns:
    if "age" in col:
        df[col] = df[col] / 10

# uzun yol
df.head()
```

```python
df[["age","age2","age3"]].apply(lambda x  : x ** 2 ).head()

# daha kısa yol ama iyi değil(değişken isimlerini manuel yazdık)
```

```python
df.loc[:, df.columns.str.contains("age")].apply(lambda x : x / 10).head()

# kısa yol ve otomatik
```

```python
df.loc[:, df.columns.str.contains("age")].apply(lambda x : (x - x.mean()) / x.std()).head()

# datayı standartlaştırdı
```

```python
def standart_scaler(colName):
    return (colName - colName.mean()) / colName.std()

df.loc[:, df.columns.str.contains("age")].apply(lambda x : standart_scaler(x)).head()

# fonksiyonu dışarıda tanımlayarak daha anlaşılır hale getirdik
```

```python
def standart_scaler(colName):
    return (colName - colName.mean()) / colName.std()

df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()

# kullanım 2
```

```python
# df.loc[: , ["age","age2","age3"]] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)

df.loc[: , df.columns.str.contains("age")] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)

# oluşturduğumuz verileri df e kaydetmek 

df.head()
```

```python
df1 = pd.DataFrame({'employees': ['john', 'dennis', 'mark', 'maria'],
                    'group': ['accounting', 'engineering', 'engineering', 'hr']})

df2 = pd.DataFrame({'employees': ['mark', 'john', 'dennis', 'maria'],
                    'start_date': [2010, 2009, 2014, 2019]})
```

```python
pd.merge(df1,df2)
```

```python
df3 = pd.merge(df1,df2, on= "employees")

# neye göre birleştireceğimizi on = var ile belirleriz
```

```python
df4 = pd.DataFrame({'group': ['accounting', 'engineering', 'hr'],
                    'manager': ['Caner', 'Mustafa', 'Berkcan']})
```

```python
pd.merge(df3,df4)
```
